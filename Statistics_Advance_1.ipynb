Q1. What is a random variable in probability theory?

Ans. Random variable consists a set of possible values from a random experiments.

Q2. What are the types of random variables?

Ans. Random variables are classified into two main types based on the nature of the outcomes they represent:

1. **Discrete Random Variable**: Takes on a countable number of distinct values. These are typically associated with scenarios where outcomes can be listed, such as the number of heads in coin tosses or the result of rolling a die. The probability distribution for a discrete random variable is given by a probability mass function (PMF).

2. **Continuous Random Variable**: Takes on an uncountably infinite number of values within a given range. These are used to model quantities that can vary smoothly, such as height, weight, or time. The probability distribution for a continuous random variable is described by a probability density function (PDF), and probabilities are calculated over intervals, not specific points.

In essence, discrete variables count outcomes, while continuous variables measure them.

Q3. What is the difference between discrete and continuous distributions?

Ans. The key difference between **discrete** and **continuous distributions** lies in the type of values their random variables can take:

1. **Discrete Distribution**: Deals with discrete random variables that assume countable, separate values (e.g., 0, 1, 2,...). The distribution is defined by a **probability mass function (PMF)**, which assigns probabilities to each possible value. The total sum of all probabilities is 1.

2. **Continuous Distribution**: Deals with continuous random variables that can take any value within a given range (e.g., all real numbers between 0 and 1). It is defined by a **probability density function (PDF)**, and probabilities are calculated over intervals using integration. The probability of the variable taking any exact value is zero.

In short, discrete distributions apply to countable outcomes with distinct probabilities, while continuous distributions apply to measurable outcomes with probabilities over intervals.

Q4. What are probability distribution functions (PDF)?

Ans. A Probability Distribution Function (PDF) refers to a function that describes the likelihood of a random variable taking on certain values. There are two types, depending on the nature of the random variable:

1. For Discrete Random Variables: The PDF is called a Probability Mass Function (PMF). It gives the probability that a discrete random variable is exactly equal to a specific value.

2. For Continuous Random Variables: The PDF refers to the Probability Density Function. It describes the relative likelihood of the variable falling within a particular range.

In summary, a PDF is a function that defines how probabilities are distributed over the values of a random variableâ€”discrete via PMF, and continuous via a density function.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

Ans. A **Cumulative Distribution Function (CDF)** differs from a **Probability Distribution Function (PDF)** in that while the PDF (or PMF for discrete variables) describes the probability or probability density at specific values of a random variable, the CDF gives the cumulative probability that the variable takes on a value less than or equal to a certain threshold. For discrete variables, the PDF is the Probability Mass Function (PMF), which assigns probabilities to exact outcomes, whereas for continuous variables, the PDF is the Probability Density Function, which defines a density over a range, and actual probabilities are calculated over intervals using integration. In contrast, the CDF applies to both types of variables and is defined as $F(x) = P(X \leq x)$, representing the total probability accumulated up to and including the value $x$. The CDF is always non-decreasing, starts at 0, and approaches 1 as $x$ approaches infinity, while the PDF or PMF focuses on the local probability behavior rather than cumulative totals.

Q8. What is the binomial distribution, and how is it used in probability?

Ans. The **binomial distribution** is a discrete probability distribution that models the number of successes in a fixed number $n$ of independent trials, where each trial has two outcomes (success or failure) and the same probability of success $p$. The probability of getting exactly $k$ successes is given by $P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k}$. It is widely used to represent real-life situations like the number of heads in coin tosses, defective items in production, or positive responses in surveys, with mean $np$ and variance $np(1 - p)$.

Q9. What is the Poisson distribution and where is it applied?

Q10. What is a continuous uniform distribution?

Ans. A **continuous uniform distribution** is a probability distribution where a continuous random variable is equally likely to take any value within a specified interval $[a, b]$. The probability density function (PDF) is constant over this interval and is given by:

$$
f(x) = \frac{1}{b - a}, \quad \text{for } a \leq x \leq b
$$

and $f(x) = 0$ outside this interval. This means the probability is spread evenly across the range, with no value more likely than another. The **mean** of the distribution is $\frac{a + b}{2}$ and the **variance** is $\frac{(b - a)^2}{12}$. It is used in scenarios where all outcomes in a range are equally likely, such as selecting a random number between two bounds or modeling arrival times within a fixed window.

Q11. What are the characteristics of a normal distribution?

Ans. A **normal distribution**, also known as a Gaussian distribution, is a continuous probability distribution with the following key characteristics:

1. **Bell-shaped curve**: It is symmetric around the mean, with the highest point at the center.
2. **Mean = Median = Mode**: All three central measures are equal and located at the center of the distribution.
3. **Defined by two parameters**: The **mean** $\mu$ (which determines the center) and the **standard deviation** $\sigma$ (which controls the spread).
4. **Empirical Rule (68-95-99.7 rule)**: About 68% of the data falls within 1 standard deviation of the mean, 95% within 2, and 99.7% within 3.
5. **Asymptotic**: The tails extend infinitely in both directions, never touching the horizontal axis.
6. **Unimodal**: It has a single peak at the mean.
7. **Probability density function (PDF)**:

   $$
   f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{- \frac{(x - \mu)^2}{2\sigma^2}}
   $$

Normal distributions are widely used in statistics and natural sciences due to the **Central Limit Theorem**, which states that the sum or average of many independent random variables tends to follow a normal distribution, regardless of their original distribution.

Q12. What is the standard normal distribution, and why is it important?

Ans. The **standard normal distribution** is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. Its probability density function (PDF) is:

$$
f(z) = \frac{1}{\sqrt{2\pi}} e^{-\frac{z^2}{2}}
$$

where $z$ represents the standardized variable. It is important because it provides a **reference distribution** that allows us to easily calculate probabilities and critical values for any normal distribution by converting values into **z-scores** using the formula:

$$
z = \frac{x - \mu}{\sigma}
$$

This standardization simplifies calculations, comparisons, and hypothesis testing, making the standard normal distribution a fundamental tool in statistics and probability.


Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Ans. The **Central Limit Theorem (CLT)** states that when you take sufficiently large random samples from any populationâ€”regardless of the populationâ€™s original distributionâ€”the distribution of the sample means will approximate a **normal distribution** with mean equal to the population mean and variance equal to the population variance divided by the sample size. This is critical in statistics because it allows us to make inferences about population parameters using the normal distribution even when the underlying data is not normally distributed. The CLT justifies many statistical methods, such as confidence intervals and hypothesis testing, by ensuring that the sampling distribution of the mean becomes approximately normal as sample size grows.

Q14. How does the Central Limit Theorem relate to the normal distribution?

Ans. The **Central Limit Theorem (CLT)** directly connects to the **normal distribution** by explaining why and how the normal distribution emerges in statistics: it states that the distribution of the sample means from any population, regardless of the populationâ€™s original shape, will tend to follow a normal distribution as the sample size becomes large. This means that even if the underlying data is skewed or irregular, the averages of sufficiently large samples will form a bell-shaped, normal curve. The CLT thus provides the theoretical foundation for using normal distribution-based methods to analyze sample means and perform statistical inference, making the normal distribution a universal model for many real-world phenomena involving averages.


Q15. What is the application of Z statistics in hypothesis testing?

Ans. **Z statistics** are used in hypothesis testing when the population variance is known or the sample size is large (usually $n > 30$). They help determine how many standard deviations a sample mean is from the population mean under the null hypothesis. By calculating the **Z-score**, which standardizes the difference between the sample statistic and the hypothesized population parameter, we can find the corresponding **p-value** or compare it to critical values from the standard normal distribution. This helps decide whether to **reject or fail to reject the null hypothesis**, making Z-tests a fundamental tool for testing hypotheses about population means or proportions when conditions for normal approximation are met.


Q16. How do you calculate a Z-score, and what does it represent?

Ans. A **Z-score** is calculated using the formula:

$$
Z = \frac{X - \mu}{\sigma}
$$

where $X$ is the value of the observation, $\mu$ is the population mean, and $\sigma$ is the population standard deviation. The Z-score represents the number of standard deviations that the observation $X$ is away from the mean. A positive Z-score means the value is above the mean, while a negative Z-score means it is below. It standardizes different data points to a common scale, allowing comparison across different datasets and enabling the use of the standard normal distribution for probability and hypothesis testing.

Q17. What are point estimates and interval estimates in statistics?

Ans. **Point estimates** are single-value statistics calculated from sample data used to estimate an unknown population parameter, such as the sample mean estimating the population mean. In contrast, **interval estimates** provide a range of values, called a confidence interval, within which the population parameter is expected to lie with a certain level of confidence (e.g., 95%). While point estimates give a specific guess, interval estimates express the uncertainty around that guess by giving a range supported by the data.

Q18. What is the significance of confidence intervals in statistical analysis?

Ans. Confidence intervals are significant in statistical analysis because they provide a range of plausible values for an unknown population parameter, giving more information than a single point estimate. They quantify the uncertainty inherent in sampling by showing the degree of confidence (e.g., 95%) that the true parameter lies within the interval. This helps researchers assess the precision and reliability of their estimates, make informed decisions, and communicate the uncertainty of results clearly, making confidence intervals a fundamental tool for inference and hypothesis testing.

Q19. What is the relationship between a Z-score and a confidence interval?

Ans. The **Z-score** and **confidence interval** are closely related because the Z-score determines how wide the confidence interval is around a sample estimate. Specifically, for a given confidence level (like 95%), the corresponding Z-score (called the **critical value**) defines how many standard deviations away from the mean the interval extends on both sides. The confidence interval is calculated as:

$$
\text{Estimate} \pm (Z_{\alpha/2} \times \text{Standard Error})
$$

where $Z_{\alpha/2}$ is the Z-score that captures the desired confidence level (e.g., about 1.96 for 95%). Thus, the Z-score sets the margin of error, controlling how confident we are that the interval contains the true population parameter.


Q20. How are Z-scores used to compare different distributions?

Ans. **Z-scores** are used to compare different distributions by standardizing values from different datasets onto a common scale with mean 0 and standard deviation 1. By converting raw scores into Z-scores, you can measure how far and in what direction a value deviates from its own distributionâ€™s mean in terms of standard deviations. This makes it possible to directly compare scores from different distributionsâ€”even if they have different means and spreadsâ€”by seeing which is relatively higher or lower within their own contexts. For example, a test score of 85 in one exam and 78 in another can be compared using their Z-scores to determine which performance was better relative to their respective class averages and variability.

Q21. What are the assumptions for applying the Central Limit Theorem?

Ans. The assumptions for applying the Central Limit Theorem are:

1. **Independence** of observations.
2. **Random sampling**.
3. **Sample size** should be large enough (typically n â‰¥ 30).
4. **Finite variance** in the population.
5. The population distribution doesn't need to be normal, but larger samples help.

Q22. What is the concept of expected value in a probability distribution?

Ans. The **expected value** is the weighted average of all possible outcomes of a random variable, where each outcome is weighted by its probability. It represents the average result you would expect after many trials.


Q23. How does a probability distribution relate to the expected outcome of a random variable?

Ans. A **probability distribution** describes how the values of a **random variable** are spread or distributed, showing the likelihood of each possible outcome.

The **expected outcome** (or **expected value**) of a random variable is the weighted average of all possible values, where each value is weighted by its probability. It provides the long-run average value of the random variable if you repeated the experiment many times.

Mathematically:

* For a **discrete random variable** $X$, the expected value $E(X)$ is:

  $$
  E(X) = \sum x_i \cdot P(x_i)
  $$

  where $x_i$ are the possible values, and $P(x_i)$ is the probability of each value.

* For a **continuous random variable** $X$, the expected value $E(X)$ is:

  $$
  E(X) = \int_{-\infty}^{\infty} x f(x) \, dx
  $$

  where $f(x)$ is the probability density function.

In short, the **probability distribution** tells you how likely different outcomes are, and the **expected value** is the average outcome you'd expect from many trials.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    if x == 1:
        return p
    else:
        return 1 - p

# Example usage:
p = 0.7  # Probability of success
print(bernoulli_pdf(1, p))  # Probability of success (1)
print(bernoulli_pdf(0, p))  # Probability of failure (0)
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Avoid division by zero
    if std_dev == 0:
        raise ValueError("Standard deviation is zero. Z-scores are undefined.")

    z_scores = (data - mean) / std_dev

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.scatter(range(len(z_scores)), z_scores, c='blue', alpha=0.6, label='Z-scores')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axhline(1, color='red', linestyle='--', linewidth=1, label='Z = Â±1')
    plt.axhline(-1, color='red', linestyle='--', linewidth=1)
    plt.axhline(2, color='orange', linestyle='--', linewidth=1, label='Z = Â±2')
    plt.axhline(-2, color='orange', linestyle='--', linewidth=1)
    plt.title('Z-Scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(50, 10, 100)  # 100 values from a normal distribution
z_scores = calculate_and_plot_z_scores(data)
```

```python
'''#Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_z_score(z):
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(8, 4))
    plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='blue')
    plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= z), color='red', alpha=0.3, label=f'Area â‰¤ Z={z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title("Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example
x = 75
mean = 70
std = 5

z = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x}: {z:.2f}")

plot_z_score(z)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Population parameters
mu = 100        # Population mean
sigma = 15      # Population std dev

# Sample data
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=40)
sample_mean = np.mean(sample)
n = len(sample)

# Z-statistic
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.2f}")
print(f"P-value = {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#Q16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import norm

# Sample data (e.g., product weights in grams)
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Confidence level
confidence = 0.95
z_critical = norm.ppf(1 - (1 - confidence) / 2)

# Margin of error
margin_error = z_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#Q17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy.stats import norm

# Step 1: Generate data from a normal distribution
np.random.seed(1)
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Compute sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # use ddof=1 for sample std
n = len(data)

# Step 3: Compute 95% confidence interval
confidence = 0.95
z_critical = norm.ppf(1 - (1 - confidence) / 2)
margin_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#18.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mean = 100
std_dev = 15

# Generate x values (range around the mean)
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

# Calculate the PDF
pdf = norm.pdf(x, loc=mean, scale=std_dev)

# Plot the PDF
plt.figure(figsize=(8, 4))
plt.plot(x, pdf, color='blue', label='Normal PDF')
plt.title(f'Normal Distribution PDF (Î¼={mean}, Ïƒ={std_dev})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.legend()
plt.show()
```

```python
#20. Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np

# Parameters of the uniform distribution
a = 0  # lower bound
b = 10 # upper bound

# Simulate random variable
np.random.seed(42)
samples = np.random.uniform(a, b, size=10000)

# Calculate expected value (sample mean)
expected_value = np.mean(samples)

print(f"Simulated Expected Value: {expected_value:.2f}")

# Theoretical expected value for Uniform(a, b) is (a + b) / 2
theoretical_ev = (a + b) / 2
print(f"Theoretical Expected Value: {theoretical_ev:.2f}")
```

```python
#21. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

# Generate two example datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=5, size=100)
data2 = np.random.normal(loc=50, scale=10, size=100)

# Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

print(f"Standard Deviation of Dataset 1: {std1:.2f}")
print(f"Standard Deviation of Dataset 2: {std2:.2f}")

# Visualization
plt.figure(figsize=(10,5))

# Histogram of data1
plt.hist(data1, bins=20, alpha=0.6, label=f'Dataset 1 (std={std1:.2f})', color='blue', density=True)
# Highlight Â±1 std range for data1
plt.axvline(np.mean(data1) - std1, color='blue', linestyle='--')
plt.axvline(np.mean(data1) + std1, color='blue', linestyle='--')

# Histogram of data2
plt.hist(data2, bins=20, alpha=0.6, label=f'Dataset 2 (std={std2:.2f})', color='orange', density=True)
# Highlight Â±1 std range for data2
plt.axvline(np.mean(data2) - std2, color='orange', linestyle='--')
plt.axvline(np.mean(data2) + std2, color='orange', linestyle='--')

plt.title('Comparison of Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import iqr

# Generate normal distribution data
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculate range (max - min)
data_range = np.ptp(data)

# Calculate interquartile range (IQR)
data_iqr = iqr(data)

print(f"Range: {data_range:.4f}")
print(f"Interquartile Range (IQR): {data_iqr:.4f}")
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data (e.g., skewed distribution)
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
data_normalized = (data - mean) / std

# Plot before and after normalization
plt.figure(figsize=(12,5))

plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='skyblue')
plt.title('Original Data Distribution')
plt.xlabel('Value')

plt.subplot(1, 2, 2)
sns.histplot(data_normalized, bins=30, kde=True, color='salmon')
plt.title('Z-score Normalized Data Distribution')
plt.xlabel('Z-score')

plt.tight_layout()
plt.show()
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calc_skew_kurtosis(size=1000, mean=0, std=1):
    # Generate normal data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # excess kurtosis by default

    return data_skewness, data_kurtosis

# Example usage
skewness, kurt = calc_skew_kurtosis()
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (excess): {kurt:.4f}")
```
