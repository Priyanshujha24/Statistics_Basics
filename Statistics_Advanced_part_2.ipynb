#Theory Questions

Q1. What is hypothesis testing in statistics ?

Ans. Hypothesis testing is a statistical method used to decide whether there is enough evidence to reject a null hypothesis about a population parameter.

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans. The **null hypothesis (H₀)** states there is no effect or difference. The **alternative hypothesis (H₁ or Ha)** states there is an effect or difference. Hypothesis testing determines which is more likely based on sample data.

Q3. What is the significance level in hypothesis testing, and why is it important?

Ans. The **significance level (α)** is the probability of rejecting the null hypothesis when it is actually true (Type I error). Commonly set at 0.05, it defines the threshold for statistical significance.

Q4. What does a P-value represent in hypothesis testing?

Ans. The **P-value** is the probability of obtaining results as extreme as the observed ones, assuming the null hypothesis is true. A smaller P-value indicates stronger evidence against the null hypothesis.

Q5. How do you interpret the P-value in hypothesis testing?

Ans. If the **P-value ≤ α** (e.g., 0.05), reject the null hypothesis — results are statistically significant.
If the **P-value > α**, fail to reject the null — not enough evidence against it.

Q6. What are Type 1 and Type 2 errors in hypothesis testing?

Ans. * **Type 1 error**: Rejecting a true null hypothesis (false positive).
* **Type 2 error**: Failing to reject a false null hypothesis (false negative).

Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Ans. * **One-tailed test**: Tests for an effect in one direction (e.g., > or <).
* **Two-tailed test**: Tests for an effect in both directions (e.g., ≠).

Q8. What is the Z-test, and when is it used in hypothesis testing?

Ans. The **Z-test** is used to test hypotheses about population means when the population standard deviation is known and the sample size is large (n ≥ 30).

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

Ans. Calculate Z-score using:

$$
Z = \frac{X - \mu}{\sigma}
$$

where:

* $X$ = observed value
* $\mu$ = population mean
* $\sigma$ = population standard deviation

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?

Ans. The **T-distribution** is a probability distribution used when the sample size is small (n < 30) and population standard deviation is unknown. It’s wider than normal distribution to account for extra uncertainty. Use it instead of normal distribution in these cases.

Q11. What is the difference between a Z-test and a T-test?

Ans. * **Z-test**: Used when population standard deviation is known and sample size is large (n ≥ 30).
* **T-test**: Used when population standard deviation is unknown and/or sample size is small (n < 30).

Q12. What is the T-test, and how is it used in hypothesis testing?

Ans. The **T-test** compares the sample mean to a population mean (or between two sample means) when population standard deviation is unknown. It helps determine if differences are statistically significant, especially with small samples.

Q13. What is the relationship between Z-test and T-test in hypothesis testing?

Ans. The T-test generalizes the Z-test:

* When population standard deviation is unknown and sample size is small, use T-test.
* When population standard deviation is known and sample size is large, T-test approaches Z-test.

Q14. What is a confidence interval, and how is it used to interpret statistical results?

Ans. A **confidence interval** is a range of values that likely contains the population parameter with a given confidence level (e.g., 95%). It shows estimate precision and uncertainty in statistical results.

Q15. What is the margin of error, and how does it affect the confidence interval?

Ans. The **margin of error** is the maximum expected difference between the sample estimate and the true population parameter. It determines the width of the confidence interval—the larger the margin, the wider the interval.

Q16. How is Bayes' Theorem used in statistics, and what is its significance?

Ans. Bayes' Theorem updates the probability of a hypothesis based on new evidence. It’s significant for combining prior knowledge with data to improve decision-making and inference.

Q17. What is the Chi-square distribution, and when is it used?

Ans. The **Chi-square distribution** is a probability distribution used mainly for testing relationships between categorical variables (e.g., in Chi-square tests) and for variance estimates in normal populations.

Q18. What is the Chi-square goodness of fit test, and how is it applied?

Ans. The **Chi-square goodness of fit test** checks if observed categorical data matches expected distribution. Calculate:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

Compare to critical value to accept or reject the fit.

Q19. What is the F-distribution, and when is it used in hypothesis testing?

Ans. The **F-distribution** is used to compare variances between two populations. It’s commonly applied in ANOVA tests and tests for equality of variances.

Q20. What is an ANOVA test, and what are its assumptions?

Ans. **ANOVA (Analysis of Variance)** tests if means of three or more groups are equal.

**Assumptions:**

1. Independence of observations
2. Normal distribution of groups
3. Equal variances (homogeneity of variance) across groups

Q21. What are the different types of ANOVA tests?

Ans. * **One-way ANOVA:** Tests one factor with multiple groups.
* **Two-way ANOVA:** Tests two factors and their interaction.
* **Repeated measures ANOVA:** Tests the same subjects under different conditions.

Q22. What is the F-test, and how does it relate to hypothesis testing?

Ans. The **F-test** compares variances between groups to see if they are significantly different. It’s used in ANOVA and regression to test overall model significance.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    if x == 1:
        return p
    else:
        return 1 - p

# Example usage:
p = 0.7  # Probability of success
print(bernoulli_pdf(1, p))  # Probability of success (1)
print(bernoulli_pdf(0, p))  # Probability of failure (0)
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Avoid division by zero
    if std_dev == 0:
        raise ValueError("Standard deviation is zero. Z-scores are undefined.")

    z_scores = (data - mean) / std_dev

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.scatter(range(len(z_scores)), z_scores, c='blue', alpha=0.6, label='Z-scores')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axhline(1, color='red', linestyle='--', linewidth=1, label='Z = ±1')
    plt.axhline(-1, color='red', linestyle='--', linewidth=1)
    plt.axhline(2, color='orange', linestyle='--', linewidth=1, label='Z = ±2')
    plt.axhline(-2, color='orange', linestyle='--', linewidth=1)
    plt.title('Z-Scores of the Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    return z_scores

# Example usage
data = np.random.normal(50, 10, 100)  # 100 values from a normal distribution
z_scores = calculate_and_plot_z_scores(data)
```

```python
'''#Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_z_score(z):
    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(8, 4))
    plt.plot(x_vals, y_vals, label='Standard Normal PDF', color='blue')
    plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= z), color='red', alpha=0.3, label=f'Area ≤ Z={z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title("Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example
x = 75
mean = 70
std = 5

z = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x}: {z:.2f}")

plot_z_score(z)
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Population parameters
mu = 100        # Population mean
sigma = 15      # Population std dev

# Sample data
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=40)
sample_mean = np.mean(sample)
n = len(sample)

# Z-statistic
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-statistic = {z:.2f}")
print(f"P-value = {p_value:.4f}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
#Q16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import norm

# Sample data (e.g., product weights in grams)
np.random.seed(0)
data = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5, n=100

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# Confidence level
confidence = 0.95
z_critical = norm.ppf(1 - (1 - confidence) / 2)

# Margin of error
margin_error = z_critical * (sample_std / np.sqrt(n))

# Confidence interval
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{int(confidence*100)}% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calc_skew_kurtosis(size=1000, mean=0, std=1):
    # Generate normal data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)  # excess kurtosis by default

    return data_skewness, data_kurtosis

# Example usage
skewness, kurt = calc_skew_kurtosis()
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (excess): {kurt:.4f}")
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import scipy.stats as stats
import math

# Sample data
sample_mean = 105
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05  # Significance level

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))

# Calculate P-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpretation
if p_value <= alpha:
    print("Reject the null hypothesis: significant difference.")
else:
    print("Fail to reject the null hypothesis: no significant difference.")
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import math
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Data
sample_mean = 55
population_mean = 50
population_std = 10
sample_size = 30
alpha = 0.05

# Calculate Z-score
z = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))
p = 2 * (1 - norm.cdf(abs(z)))

print(f"Z-score: {z:.3f}")
print(f"P-value: {p:.3f}")

if p <= alpha:
    print("Reject null hypothesis: significant difference.")
else:
    print("Fail to reject null hypothesis: no significant difference.")

# Plot
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Critical Z values for two-tailed test
z_crit = norm.ppf(1 - alpha/2)

plt.plot(x, y, label='Normal Distribution')
plt.fill_between(x, 0, y, where=(x <= -z_crit), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, 0, y, where=(x >= z_crit), color='red', alpha=0.3)

plt.axvline(z, color='blue', linestyle='--', label=f'Z-score = {z:.2f}')
plt.axvline(-z_crit, color='red', linestyle=':')
plt.axvline(z_crit, color='red', linestyle=':')

plt.title('Two-tailed Z-test with Rejection Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import math

def plot_type1_type2_errors(mu0, mu1, sigma, n, alpha):
    """
    mu0: population mean under H0
    mu1: true population mean under H1
    sigma: population standard deviation
    n: sample size
    alpha: significance level (Type 1 error rate)
    """

    # Calculate critical Z for alpha
    z_crit = norm.ppf(1 - alpha)
    # Critical sample mean value to reject H0
    crit_val = mu0 + z_crit * (sigma / math.sqrt(n))

    # Define range for sample means
    x = np.linspace(mu0 - 4*sigma/np.sqrt(n), mu1 + 4*sigma/np.sqrt(n), 1000)

    # Distributions under H0 and H1
    h0_pdf = norm.pdf(x, mu0, sigma/np.sqrt(n))
    h1_pdf = norm.pdf(x, mu1, sigma/np.sqrt(n))

    # Plot distributions
    plt.plot(x, h0_pdf, label='H0 Distribution (mu=%.2f)' % mu0)
    plt.plot(x, h1_pdf, label='H1 Distribution (mu=%.2f)' % mu1)

    # Shade Type 1 error region (rejecting H0 when true)
    x_type1 = np.linspace(crit_val, x[-1], 1000)
    plt.fill_between(x_type1, 0, norm.pdf(x_type1, mu0, sigma/np.sqrt(n)), color='red', alpha=0.3, label='Type 1 Error (α)')

    # Shade Type 2 error region (failing to reject H0 when false)
    x_type2 = np.linspace(x[0], crit_val, 1000)
    plt.fill_between(x_type2, 0, norm.pdf(x_type2, mu1, sigma/np.sqrt(n)), color='blue', alpha=0.3, label='Type 2 Error (β)')

    plt.axvline(crit_val, color='black', linestyle='--', label='Critical Value')

    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

    # Calculate Type 2 error probability (beta)
    beta = norm.cdf(crit_val, mu1, sigma / math.sqrt(n))
    power = 1 - beta

    print(f"Type 1 error rate (α): {alpha}")
    print(f"Type 2 error rate (β): {beta:.4f}")
    print(f"Statistical Power (1 - β): {power:.4f}")

# Example usage:
plot_type1_type2_errors(mu0=100, mu1=105, sigma=15, n=36, alpha=0.05)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

# Sample data (two independent groups)
group1 = np.array([23, 20, 22, 21, 19, 24, 25])
group2 = np.array([30, 28, 26, 27, 29, 31, 33])

# Perform independent T-test
t_stat, p_value = ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

# Interpretation at alpha = 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: groups have significantly different means.")
else:
    print("Fail to reject the null hypothesis: no significant difference between groups.")
```

```python
#7.Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt

# Sample paired data (e.g., before and after treatment)
before = np.array([85, 88, 90, 86, 87, 89, 91, 84])
after = np.array([88, 90, 93, 89, 90, 92, 94, 87])

# Perform paired T-test
t_stat, p_value = ttest_rel(before, after)

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.3f}")

alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: significant difference between paired samples.")
else:
    print("Fail to reject null hypothesis: no significant difference.")

# Visualization
plt.figure(figsize=(8,5))
plt.plot(range(len(before)), before, 'o-', label='Before')
plt.plot(range(len(after)), after, 's-', label='After')
plt.xticks(range(len(before)), [f'Subject {i+1}' for i in range(len(before))])
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Paired Sample Comparison')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#8.Simulate data and perform both Z-test and T-test, then compare the results using Python.

import numpy as np
import math
from scipy.stats import norm, t

# Parameters
population_mean = 100
population_std = 15  # known for Z-test
sample_size = 25
alpha = 0.05

# Simulate sample data
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=sample_size)
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)

# --- Z-test ---
z_score = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))
z_p_value = 2 * (1 - norm.cdf(abs(z_score)))

# --- T-test ---
t_score = (sample_mean - population_mean) / (sample_std / math.sqrt(sample_size))
t_p_value = 2 * (1 - t.cdf(abs(t_score), df=sample_size - 1))

# Results
print("Sample Mean:", round(sample_mean, 2))
print("\n--- Z-test ---")
print(f"Z-score: {z_score:.3f}")
print(f"P-value: {z_p_value:.3f}")
print("Conclusion:", "Reject H0" if z_p_value < alpha else "Fail to reject H0")

print("\n--- T-test ---")
print(f"T-score: {t_score:.3f}")
print(f"P-value: {t_p_value:.3f}")
print("Conclusion:", "Reject H0" if t_p_value < alpha else "Fail to reject H0")

# Comparison Summary
if (z_p_value < alpha) == (t_p_value < alpha):
    print("\n✅ Both tests lead to the same conclusion.")
else:
    print("\n⚠️ Z-test and T-test lead to different conclusions. (Check sample size & variance assumptions.)")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for the sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_crit * std_err

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return round(lower, 2), round(upper, 2)

# Example usage:
sample_data = [50, 52, 48, 49, 51, 50, 47]
ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean.

    Parameters:
    - data: list or array of sample values
    - confidence: desired confidence level (default is 0.95 for 95%)

    Returns:
    - Margin of error (rounded to 2 decimals)
    """
    n = len(data)
    std_dev = np.std(data, ddof=1)
    std_err = std_dev / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_crit * std_err
    return round(moe, 2)

# Example usage
sample_data = [72, 75, 70, 74, 73, 71, 76]
confidence_level = 0.95

moe = margin_of_error(sample_data, confidence=confidence_level)
print(f"Margin of Error at {int(confidence_level*100)}% confidence level: ±{moe}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_H, likelihood_E_given_H, likelihood_E_given_not_H):
    """
    Apply Bayes' Theorem to update the probability of a hypothesis.

    Parameters:
    - prior_H: Prior probability of the hypothesis (P(H))
    - likelihood_E_given_H: Likelihood of evidence if hypothesis is true (P(E|H))
    - likelihood_E_given_not_H: Likelihood of evidence if hypothesis is false (P(E|¬H))

    Returns:
    - Posterior probability (P(H|E))
    """
    prior_not_H = 1 - prior_H
    marginal_E = (likelihood_E_given_H * prior_H) + (likelihood_E_given_not_H * prior_not_H)
    posterior_H = (likelihood_E_given_H * prior_H) / marginal_E
    return round(posterior_H, 4)

# Example scenario:
# Disease test: 1% of population has the disease
# Test is 99% accurate when disease is present, 95% accurate when absent

prior = 0.01                 # P(Disease)
likelihood_positive_given_disease = 0.99  # P(Pos | Disease)
likelihood_positive_given_no_disease = 0.05  # P(Pos | No Disease)

posterior = bayes_theorem(prior, likelihood_positive_given_disease, likelihood_positive_given_no_disease)
print(f"Posterior probability (Disease | Positive test): {posterior * 100:.2f}%")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
from scipy.stats import chi2_contingency

# Example contingency table
# Rows = Gender (Male, Female)
# Columns = Preference (Product A, Product B)
data = np.array([
    [30, 10],   # Male
    [20, 40]    # Female
])

# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

print(f"Chi-square statistic: {chi2_stat:.3f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: Variables are dependent.")
else:
    print("Fail to reject null hypothesis: Variables are independent.")
```

```python
#14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

from scipy.stats import chisquare
import numpy as np

# Example observed data (e.g., frequencies of outcomes)
observed = np.array([18, 22, 20, 25, 15])

# Expected distribution (e.g., uniform or custom)
expected = np.array([20, 20, 20, 20, 20])  # uniform distribution

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output
print("Observed:", observed)
print("Expected:", expected)
print(f"\nChi-square Statistic: {chi2_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: observed data does NOT fit expected distribution.")
else:
    print("Fail to reject null hypothesis: observed data fits expected distribution.")
```
